```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
from extract_patches.core import extract_patches

```

```python
img1 = cv2.imread("kpi.png",0)
det = cv2.AKAZE_create();
kps = det.detect(img1,None)
```

```python
patches = extract_patches(kps, img1, 64, 12.0)
```

```python
for i in range(5):
    plt.figure()
    python = patches[i].astype(np.uint8)
    cpp = cv2.imread(f'{i}.png', 0).astype(np.uint8)
    plt.imshow(np.concatenate([python, cpp],axis=1))
               
```
